```python
# Inline Display
%matplotlib inline
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to months as stings
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set y-axis to total fare in US dollars as float values
y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create plot
plt.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# New notation
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create plot with label
plt.plot(x_axis, y_axis, marker ="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set y lim
plt.ylim(0,45)
# Set title
plt.title("PyBer Fare by Month")
# Add grid
plt.grid()
# Add legend
plt.legend()
```

```python
# Create plot with label, color, marker, and linewidth
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color = "green",marker = "D", linewidth=2, label="Boston")
# Add legend
ax.legend
# Add title
ax.set_title("PyBer Fare by Month")
# Add labels for axes
ax.set_xlabel("Month")
ax.set_ylabel("Fares($)")
# Add grid
ax.grid()
```

```python
# Plot scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Plot scatter plot
plt.scatter(x_axis, y_axis)
```

```python
# Scatter plot practice
plt.scatter(y_axis, x_axis, label="Chicago", color="Red")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
# Plot scatter with sizing
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Magnify size differences
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
# PLot new scatter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Condense code with list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Object oriented scatter plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Object oriented scatter plot with formatting
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           color="skyblue",  
           s = [i * 5 for i in y_axis], 
           alpha = 0.8, 
           edgecolors="black", 
           linewidth = 2, 
           label="Boston")
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,50)
plt.gca().invert_yaxis()
```

```python
# Plot pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Format pie chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%")
plt.show()
```

```python
# Reassign colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct="%.1f%%")
plt.show()
```

```python
# Pie chart using object oriented interface
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Pie practice
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, labels=x_axis,
       autopct="%.1f%%",
       explode = explode_values,
      startangle = 90,
      shadow=True,
       counterclock=False,
       colors=colors)
plt.show()
```
